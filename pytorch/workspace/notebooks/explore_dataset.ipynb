```python
data_dir = "/home/datasets/indoor_train/train/HR"
data_dict = {}
```

```python
for root in os.listdir(data_dir):
    location_dir = os.path.join(data_dir, root)
    for image in os.listdir(os.path.join(location_dir, "color")):
        base_image = image[:-5]
        color_image = os.path.join(location_dir, "color") + "/" + image
        depth_filled = os.path.join(location_dir, "depth_filled") + "/" + base_image + "depth_filled.png"
        data_dict[base_image] = {"color": color_image, "depth": depth_filled}
```

```python
from random import randrange

def random_squarecropper(image):
    """
    Crops a 20-50 pixel square randomly throughout an image.
    """
    image_size = image.shape
    x_size = image_size[0]
    y_size = image_size[1]
    
    x_pos = randrange(x_size)
    y_pos = randrange(y_size)
    
    square_size = randrange(50,200)
    
    image[x_pos:x_pos+square_size, y_pos:y_pos+square_size, :] = 0
    
    return image
```
